```python
import os

def read_file(file_path):
    """
    Reads a text file and returns its content as a list of lines.
    """
    with open(file_path, 'r', encoding='utf-8-sig') as file:
        return file.readlines()

def compare_file_lengths(lines1, lines2):
    """
    Compares the lengths of two lists of lines and returns a difference message if they are different.
    """
    len1 = len(lines1)
    len2 = len(lines2)
    if len1 != len2:
        return f"File lengths differ: File 1 has {len1} lines, File 2 has {len2} lines.\n"
    return ""

def compare_line_format(line1, line2):
    """
    Compares the format of two lines. Returns a message if formats differ.
    """
    def is_valid_format(line):
        parts = line.split(',')
        if len(parts) < 9:
            return False
        points = parts[:8]
        if all(part.strip().lstrip('-').isdigit() for part in points):
            return True
        return False

    if is_valid_format(line1) and is_valid_format(line2):
        return ""
    return f"Format mismatch:\nFile 1: {line1}\nFile 2: {line2}\n"

def compare_files(file1_path, file2_path, output_path):
    """
    Compares the contents of two files by length and line format, writing differences to an output file.
    """
    lines1 = read_file(file1_path)
    lines2 = read_file(file2_path)

    differences = []

    # Compare file lengths
    length_diff = compare_file_lengths(lines1, lines2)
    if length_diff:
        differences.append(length_diff)

    # Compare line formats
    max_lines = max(len(lines1), len(lines2))
    for i in range(max_lines):
        line1 = lines1[i].strip() if i < len(lines1) else ""
        line2 = lines2[i].strip() if i < len(lines2) else ""

        if line1 != line2:
            format_diff = compare_line_format(line1, line2)
            if format_diff:
                differences.append(f"Line {i + 1}:\n{format_diff}")

    # Write differences to output file
    with open(output_path, 'w', encoding='utf-8') as output_file:
        if differences:
            output_file.writelines(differences)
        else:
            output_file.write("The files are identical in length and format.\n")

# Example usage
folder_path = "ing/Detection_Dataset/check"
file1_name = "gt_img_107.txt"
file2_name = "gt_img_108.txt"
output_name = "result.txt"

file1_path = os.path.join(folder_path, file1_name)
file2_path = os.path.join(folder_path, file2_name)
output_path = os.path.join(folder_path, output_name)

compare_files(file1_path, file2_path, output_path)
```
